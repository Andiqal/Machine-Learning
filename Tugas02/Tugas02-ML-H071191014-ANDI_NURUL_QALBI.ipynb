```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from pandas import DataFrame
```

```python
n = 1000
student_data = pd.DataFrame({
    "berat kg": np.random.randint(40, 80, n),
    "tinggi cm": np.random.randint(150, 190, n)
})
```

```python
student_data
```

```python
kmeans = KMeans(5)
kmeans.fit(student_data)
```

```python
student_data["cluster"] = kmeans.labels_
```

```python
student_data
```

```python
student_data.plot.scatter('berat kg','tinggi cm', c='cluster', colormap='gist_rainbow')
```
